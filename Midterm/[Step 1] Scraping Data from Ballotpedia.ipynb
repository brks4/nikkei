# Scraping data from Ballotpedia

### List of candidates running in U.S. Congress elections, 2018
https://ballotpedia.org/List_of_candidates_running_in_U.S._Congress_elections,_2018

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
response = requests.get('https://ballotpedia.org/List_of_candidates_running_in_U.S._Congress_elections,_2018')
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
states = doc.find_all('h1')[1:]
```

```python
chambers = doc.find_all('h2')[5:88]
```

```python
parties = doc.find_all('h3')[1:250]
```

```python
candidates = doc.find_all('div')[58:307]
```

```python
candidates
```

```python
chamber_list = []
for chamber in chambers:
    chamber_list.append(chamber.text)
    chamber_list.append(chamber.text)
    chamber_list.append(chamber.text)
```

```python
party_list = []
for party in parties:
    party_list.append(party.text)
```

```python
candidate_list = []
for candidate in candidates:
    candidate_list.append(candidate)
```

```python
d = {'chamber': chamber_list, 'party': party_list, 'candidate': candidate_list}
```

```python
df = pd.DataFrame(d)
```

```python
df.head()
```

```python
df_house = df[df.chamber == 'U.S. House'].reset_index(drop=True)
```

```python
state_list = []
for state in states:
    state_list.append(state.text)
    state_list.append(state.text)
    state_list.append(state.text)    
```

```python
df_house['state'] = state_list
```

```python
df_house = df_house[df_house.party != 'Third Party'].reset_index(drop=True)
```

```python
df_house.head()
```

```python
candidate_list = []
for i in range(0,len(df_house)):
    for candidate in df_house.candidate[i].find_all('li'):
        candidate_dic = {}
        candidate_dic['party'] = df_house.party[i]
        candidate_dic['state'] = df_house.state[i]        
        candidate_dic['name'] = candidate.text
        candidate_dic['url'] = 'https://ballotpedia.org'+candidate.find('a').get('href')
        candidate_list.append(candidate_dic)
```

```python
candidate_df = pd.DataFrame(candidate_list)
```

```python
candidate_df.head()
```

```python
candidate_df.to_csv('ballotpedia.csv',index=False)
```

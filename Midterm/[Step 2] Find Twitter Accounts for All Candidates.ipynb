### List of candidates running in U.S. Congress elections, 2018
https://ballotpedia.org/List_of_candidates_running_in_U.S._Congress_elections,_2018

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import time
```

```python
df = pd.read_csv('ballotpedia.csv')
```

```python
df.head()
```

```python
len(df)
```

```python
candidate_list = []
for i in range(0,len(df)):
    response = requests.get(df.url[i])
    doc = BeautifulSoup(response.text, 'html.parser')
    candidate_dic = {}
    candidate_dic['name'] = df.name[i]
    candidate_dic['url'] = df.url[i]
    candidate_dic['state'] = df.state[i]
    candidate_dic['party'] = df.party[i]    
    try:
        candidate_dic['district'] = doc.find(class_='infobox person').find('p').text
        candidate_dic['district_alt'] = doc.find(class_='infobox person').find_all('p')[1].text        
        for link in doc.find(class_='infobox person').find_all('a'):
            if link.text == 'Twitter':
                candidate_dic['twitter'] = link.get('href')
    except:
        pass
    candidate_list.append(candidate_dic)
    print(str(i)+'/'+str(len(df)-1), df.name[i])
    time.sleep(1)
```

# Verifing twitter account

```python
label_list = []
for i in range(0,len(twitter_df)):
    response = requests.get(twitter_df.twitter[i])
    doc = BeautifulSoup(response.text, 'html.parser')
    label_dic = {}
    label_dic['name'] = twitter_df.name[i]
    label_dic['state'] = twitter_df.state[i]
    label_dic['district'] = twitter_df.district_N[i]
    try:
        label_dic['label'] = doc.find(class_='HighlightedUserLabel-description').text
    except:
        pass
    label_list.append(label_dic)
    print(str(i)+'/'+str(len(twitter_df)-1),twitter_df.name[i])
    time.sleep(1)
```

```python
label_df = pd.DataFrame(label_list)
```

```python
label_df.head()
```

```python
label_df = label_df[~label_df.label.isnull()].reset_index(drop=True)
```

```python
label_df = label_df[label_df.label.str.contains('US House candidate')].reset_index(drop=True)
```

```python
label_df['abbreviation'] = label_df.label.str.extract(', ([A-Z]+)')
```

```python
label_df['district_N'] = label_df.label.str.extract('(\d+)')
```

```python
label_df.district_N = label_df.district_N.fillna('0')
```

```python
label_df.district = label_df.district.astype(str)
```

```python
label_df['abb'] = np.nan
```

```python
label_df.head()
```

```python
abb_df = pd.read_csv('states_abb.csv')
```

```python
abb_df.head()
```

```python
for i in range(0,len(label_df)):
    for j in range(0,len(abb_df)):
        if label_df.iloc[i,3] == abb_df.iloc[j,0]:
            label_df.iloc[i,6] = abb_df.iloc[j,1]
```

```python
label_df.head()
```

```python
label_df = label_df[(label_df.abbreviation == label_df.abb) & (label_df.district == label_df.district_N)]
```

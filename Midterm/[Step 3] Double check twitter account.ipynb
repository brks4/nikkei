# Double check twitter account

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import time
```

```python
house_df = pd.read_csv('house_candidates.csv')
```

```python
house_df.head()
```

```python
twitter_df = pd.read_csv('twitter_account_verified.csv')
```

```python
twitter_df.head()
```

```python
df_to_check = house_df[~house_df.name.isin(twitter_df.name)].reset_index(drop=True)
```

```python
df_to_check.head()
```

```python
len(df_to_check)
```

# Search "Campaign Twitter"

```python
candidate_list = []
for i in range(0,len(df_to_check)):
    response = requests.get(df_to_check.url[i])
    doc = BeautifulSoup(response.text, 'html.parser')
    candidate_dic = {}
    candidate_dic['name'] = df_to_check.name[i]
    candidate_dic['url'] = df_to_check.url[i]
    candidate_dic['state'] = df_to_check.state[i]
    candidate_dic['party'] = df_to_check.party[i]    
    try:
        candidate_dic['district'] = doc.find(class_='infobox person').find('p').text
        candidate_dic['district_alt'] = doc.find(class_='infobox person').find_all('p')[1].text        
        for link in doc.find(class_='infobox person').find_all('a'):
            if link.text == 'Campaign Twitter':
                candidate_dic['twitter'] = link.get('href')
    except:
        pass
    candidate_list.append(candidate_dic)
    print(str(i)+'/'+str(len(df_to_check)-1), df_to_check.name[i])
    time.sleep(1)
```

# Search "Official Twitter"

```python
for i in range(0,len(df_to_check)):
    response = requests.get(df_to_check.url[i])
    doc = BeautifulSoup(response.text, 'html.parser')
    candidate_dic = {}
    candidate_dic['name'] = df_to_check.name[i]
    candidate_dic['url'] = df_to_check.url[i]
    candidate_dic['state'] = df_to_check.state[i]
    candidate_dic['party'] = df_to_check.party[i]    
    try:
        candidate_dic['district'] = doc.find(class_='infobox person').find('p').text
        candidate_dic['district_alt'] = doc.find(class_='infobox person').find_all('p')[1].text        
        for link in doc.find(class_='infobox person').find_all('a'):
            if link.text == 'Official Twitter':
                candidate_dic['twitter'] = link.get('href')
    except:
        pass
    candidate_list.append(candidate_dic)
    print(str(i)+'/'+str(len(df_to_check)-1), df_to_check.name[i])
    time.sleep(1)
```

```python
df = pd.DataFrame(candidate_list)
```

```python
df = df[~df.twitter.isnull()].reset_index(drop=True)
```

```python
df.head()
```

```python
df.district = df.district.str.extract('\t+(.*)\n')
df.district_alt = df.district_alt.str.extract('\t+(.*)\n')
```

```python
df['district_N'] = df.district.str.extract('(\d+)')
```

```python
df[df.district.isnull()]
```

```python
df.iloc[19,0] = df.iloc[19,1]
df.iloc[20,0] = df.iloc[20,1]
```

```python
df = df.drop(columns=['district_alt'])
```

```python
df.head()
```

# Verifing twitter account

```python
label_list = []
for i in range(0,len(df)):
    response = requests.get(df.twitter[i])
    doc = BeautifulSoup(response.text, 'html.parser')
    label_dic = {}
    label_dic['name'] = df.name[i]
    label_dic['state'] = df.state[i]
    label_dic['district'] = df.district_N[i]
    try:
        label_dic['label'] = doc.find(class_='HighlightedUserLabel-description').text
    except:
        pass
    label_list.append(label_dic)
    print(str(i)+'/'+str(len(df)-1),df.name[i])
    time.sleep(1)
```

```python
label_df = pd.DataFrame(label_list)
```

```python
label_df.head()
```

```python
label_df = label_df[~label_df.label.isnull()].reset_index(drop=True)
```

```python
label_df = label_df[label_df.label.str.contains('US House candidate')].reset_index(drop=True)
```

```python
label_df['abbreviation'] = label_df.label.str.extract(', ([A-Z]+)')
```

```python
label_df['district_N'] = label_df.label.str.extract('(\d+)')
```

```python
label_df.district_N = label_df.district_N.fillna('0')
```

```python
label_df.district = label_df.district.astype(str)
```

```python
label_df['abb'] = np.nan
```

```python
label_df.head()
```

```python
abb_df = pd.read_csv('states_abb.csv')
```

```python
abb_df.head()
```

```python
for i in range(0,len(label_df)):
    for j in range(0,len(abb_df)):
        if label_df.iloc[i,3] == abb_df.iloc[j,0]:
            label_df.iloc[i,6] = abb_df.iloc[j,1]
```

```python
label_df.head()
```

```python
label_df = label_df[(label_df.abbreviation == label_df.abb) & (label_df.district == label_df.district_N)]
```

```python
label_df.head()
```

```python
len(label_df)
```

```python
df = df[df.name.isin(label_df.name)].reset_index(drop=True)
```

```python
df.head()
```

```python
twitter_df = twitter_df.append(df).reset_index(drop=True)
```

# Tweepy analysis

```python
import datetime
import time

import tweepy
import pandas as pd
import numpy as np

consumer_key = 'XXXXXX'
consumer_secret = 'XXXXXX'
access_token = 'XXXXXX'
access_token_secret = 'XXXXXX'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

```python
df = pd.read_csv('twitter_account_verified_final.csv')
df['twitter'] = df.twitter.str.extract('twitter.com/(.*)')
df.head()
```

```python
len(df)
```

```python
tweet_df = pd.DataFrame(columns=['created_at','retweet_count','favorite_count','text','name','party'])
```

```python
for i in range(0,len(df)):
    tweet = []
    for status in tweepy.Cursor(api.user_timeline, screen_name='@'+df.twitter[i]).items():
        tweet.append(status._json)
    status_df = pd.DataFrame(tweet,columns=['created_at','retweet_count','favorite_count','text'])
    status_df['name'] = df.name[i]
    status_df['party'] = df.party[i]
    tweet_df = tweet_df.append(status_df)
    print(str(i)+'/'+str(len(df)-1),df.name[i])
    time.sleep(75)
```

## Followers count

```python
twitter_count = []
for candidate in df.twitter:
    try:
        user = api.get_user('@'+candidate)
        user_dic ={}
        user_dic['name'] = user.name
        user_dic['followers'] = user.followers_count
    except:
        pass
    twitter_count.append(user_dic)
```

```python
followers_df = pd.DataFrame(twitter_count)
```

```python
followers_df.head()
```

```python
followers_df = followers_df.sort_values(by='followers',ascending=False)
```

```python
followers_df.to_csv('followers_count',index=False)
```

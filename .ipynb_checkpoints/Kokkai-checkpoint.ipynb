```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('http://www.shugiin.go.jp/internet/itdb_gian.nsf/html/gian/menu.htm')
```

```python
dropdown = Select(driver.find_element_by_xpath('//*[@id="mainlayout"]/form/select'))
dropdown.select_by_visible_text('第196回国会（常会）')
button = driver.find_element_by_xpath('//*[@id="mainlayout"]/form/input')
driver.execute_script("arguments[0].scrollIntoView(true)", button)
button.click()
```

```python
results = driver.find_elements_by_tag_name('table')
```

```python
all_gian = []
for result in results[:3]:
    cases = result.find_elements_by_tag_name('tr')
    for case in cases[1:]:
        gian = {}
        kenmei = case.find_elements_by_tag_name('td')[2]
        gian['KENMEI'] = kenmei.text
        status = case.find_elements_by_tag_name('td')[3]
        gian['STATUS'] = status.text
        all_gian.append(gian)
```

```python
import pandas as pd
df = pd.DataFrame(all_gian)
df.to_csv('gian.csv', index=False)
```

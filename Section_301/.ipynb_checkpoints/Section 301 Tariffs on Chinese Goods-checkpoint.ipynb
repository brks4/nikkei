## Tariffs on Chinese Goods

#### Statement By U.S. Trade Representative Robert Lighthizer on Section 301 Action - USTR

https://ustr.gov/about-us/policy-offices/press-office/press-releases/2018/july/statement-us-trade-representative

#### The list of proposed tariffs on \$200 billion of Chinese imports - USTR

https://ustr.gov/sites/default/files/301/2018-0026%20China%20FRN%207-10-2018_0.pdf

Download the PDF file and read it.

```python
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler
import pdftotext
import re

%matplotlib inline
```

```python
def pdf_to_string(file):
    pdf_string = ""
    with open(file, "rb") as f:
        pdf = pdftotext.PDF(f)
        for page in pdf:
            pdf_string += page
    return pdf_string
```

```python
text = pdf_to_string('2018-0026 China FRN 7-10-2018_0.pdf')
lines = text.split('\n')
lines[500:520]
```

```python
item_list = []
for line in lines:
    try:
        item = re.findall(r'[\d]{4}.[\d]{2}.[\d]{2}.*', line)[0]
        item_list.append(item)
    except:
        pass
item_list[:10]
```

#### The latest 6,031 Chinese products targeted by US retaliatory tariffs in one giant table - Quartz

https://qz.com/1326053/all-6031-chinese-products-impacted-by-the-latest-200-billion-in-us-tariffs-in-one-giant-table/

Quartz said 6031 items, but actually 6037?

```python
len(item_list)
```

```python
pd.set_option('max_colwidth',-1)
df_item = pd.DataFrame(item_list)
df_item['Code'] = df_item[0].str.extract('([\d]{4}.[\d]{2}.[\d]{2})')
df_item['Product'] = df_item[0].str.extract('[\d]{4}.[\d]{2}.[\d]{2}\s+(.*)')
df = df_item[['Code','Product']]
df.to_csv('tariff.csv',index=False)
df.head()
```

```python
df['HTS-8'] = df.Code.str.replace('.', '')
df['HTS-4'] = df['HTS-8'].str.extract('(^[\d]{4})')
df.head()
```

```python
df['HTS-8'].to_csv('HTS-8.csv', index=False)
```

Now, we get 8-digit code for each product. By using these codes, we can get import data from USITC database.

### Go to another notebook: USITC's DataWeb

After you get the result, read `HTS.csv`.

```python
df_import = pd.read_csv('HTS.csv', dtype=str)
df_import['IMPORT'] = df_import['IMPORT'].str.replace(',','')
df_import['IMPORT'] = df_import['IMPORT'].astype(int)
df_import.sort_values(by=['IMPORT'],ascending=False).head()
```

But item lists are 5000 lines and too long. It is convenient to aggregate into larger category such as `HTS-4`.

```python
df_import['HTS-4'] = df_import['HTS Number'].str.extract('(^[\d]{4})')
df_import.head()
```

```python
df_HTS = df_import.groupby(by=['HTS-4'])['IMPORT'].sum().reset_index()
df_HTS['HTS-4'].to_csv('HTS-4.csv',index=False)
df_HTS.sort_values(by='IMPORT', ascending=False).head()
```

```python
df_commodities = pd.read_csv('COMMODITIES.csv', dtype=str)
df_commodities = df_commodities.sort_values(by='HTS-4').reset_index().drop(columns=['index'])
df_commodities.head()
```

```python
df_merge = pd.merge(df_commodities, df_HTS, on=['HTS-4','HTS-4'])
df_merge.sort_values(by='IMPORT',ascending=False).head(10)
```

```python
plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'axes.grid': True,
    'axes.prop_cycle': cycler('color', ['#01579B', '#B71C1C', '#827717', '#1B5E20', '#E65100', '#311B92', '#263238']),
    'axes.edgecolor': '#ffffff',
    'xtick.major.size': 0,
    'xtick.minor.size': 0,
    'ytick.major.size': 0,
    'ytick.minor.size': 0,
    'legend.edgecolor': '#ffffff',
    'ps.fonttype': 42,
})
```

```python
ax = df_merge.sort_values(by='IMPORT').tail(10).plot(kind='barh', x='HTS-4', y='IMPORT')
ax.set_title('Top 10 imports from China in the tariff list')
ax.grid(False, axis='y')
ax.legend_.remove()
plt.show()
```
